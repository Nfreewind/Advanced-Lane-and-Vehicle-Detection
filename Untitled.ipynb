```python
from matplotlib import pyplot as plt
import numpy as np
import scipy.misc
import cv2
import csv
from random import random
```

```python
def add_noise(x):
        return random_noise(x, mode='gaussian')

def get_image_and_steering(path,steering):
    image = scipy.misc.imresize(scipy.misc.imread(path)[25:135], [66, 200])

    if(coin_flip()):
        image = random_gamma_correction_rgb(image)

    if(coin_flip()):
        image = random_brightness_change_rgb(image)

    if(coin_flip()):
        image = random_saturation_change(image)

    if(coin_flip()):
        image = random_lightness_change(image)

    image = random_shadow(image)

    if(coin_flip()):
        image = random_blur(image)

    if(coin_flip()):
        image, steering = random_translation(image,steering)

    if(coin_flip()):
        image, steering = horizontal_flip_image(image,steering)

    return (image/255.0)-0.5, steering

def coin_flip():
    return random()<0.5

def make_yuv_grey_scale(x):
    x = np.array(x)
    x[:,:,1] = 0
    x[:,:,2] = 0
    return x

def random_gamma_correction_rgb(x):
    # Partially taken from http://www.pyimagesearch.com/2015/10/05/opencv-gamma-correction/
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    gamma = 0.4 + random() * 1.2
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")

    # apply gamma correction using the lookup table
    return cv2.LUT(x, table)

def random_brightness_change_rgb(x):
    brightness_change = 0.4 + random()*1.2
    x = np.array(x)
    x = cv2.cvtColor(x,cv2.COLOR_RGB2HSV)
    x[:,:,2] = x[:,:,2]*brightness_change
    return cv2.cvtColor(x,cv2.COLOR_HSV2RGB)


def random_saturation_change(x):
    saturation_change = 0.4 + 1.2*random()
    x = np.array(x)
    x = cv2.cvtColor(x,cv2.COLOR_RGB2HSV)
    x[:,:,1] = x[:,:,1]*saturation_change
    return cv2.cvtColor(x,cv2.COLOR_HSV2RGB)

def random_lightness_change(x):
    lightness_change = 0.4 + 1.2*random()
    x = np.array(x)
    x = cv2.cvtColor(x,cv2.COLOR_RGB2HLS)
    x[:,:,1] = x[:,:,1]*lightness_change
    return cv2.cvtColor(x,cv2.COLOR_HLS2RGB)


def random_translation(x,steer):
    x = np.array(x)
    rows,cols,rgb = x.shape

    rand_for_x = random()

    translate_y = -5 + random()*10
    translate_x = -30 + rand_for_x*60

    M = np.float32([[1,0,translate_x],[0,1,translate_y]])
    return cv2.warpAffine(x,M,(cols,rows)), (steer+(rand_for_x-0.5)*0.2)

def random_rotation_image(x):
    x = np.array(x)
    rows,cols,rgb = x.shape

    rand_angle = 3*(random()-0.5)

    M = cv2.getRotationMatrix2D((cols/2,rows/2),rand_angle,1)
    x = cv2.warpAffine(x,M,(cols,rows))
    return x

def horizontal_flip_image(x,steer):
    steer = -steer
    x = np.array(x)
    return cv2.flip(x,1), steer


def random_shadow(x):
    x = cv2.cvtColor(x,cv2.COLOR_RGB2HSV)

    max_x = 200
    max_y = 66

    if(coin_flip()):
        i_1 = (0,0)
        i_2 = (0,max_y)
        i_3 = (random()*max_x,max_y)
        i_4 = (random()*max_x,0)
    else:
        i_1 = (random()*max_x,0)
        i_2 = (random()*max_x,max_y)
        i_3 = (max_x,max_y)
        i_4 = (max_x,0)

    vertices = np.array([[i_1,i_2,i_3,i_4]], dtype = np.int32)

    x = region_of_interest(x,vertices)

    x = cv2.cvtColor(x,cv2.COLOR_HSV2RGB)
    return x

def random_blur(x):
    kernel_size = 1+int(random()*9)
    if(kernel_size%2 == 0):
        kernel_size = kernel_size + 1
    x = cv2.GaussianBlur(x,(kernel_size,kernel_size),0)
    return x

def region_of_interest(x, vertices):

    random_brightness = 0.3 + random()*1.4
    random_saturation = 0.3 + random()*1.4
    mask = np.zeros_like(x)

    ignore_mask_color = [0,0,255]

    cv2.fillPoly(mask, vertices, ignore_mask_color)

    indices = mask[:,:,2] == 255

    x[:,:,2][indices] = x[:,:,2][indices]*random_brightness
    x[:,:,1][indices] = x[:,:,1][indices]*random_saturation

    return x
```

```python
metadata = []

filename = 'driving_log.csv'

# loading metadata
with open(filename, 'r') as f:
    reader = csv.reader(f)
    i = 0
    for row in reader:
        metadata.append(row)

# removing first row if it has column names
if(metadata[0][0]=='center'):
    metadata.reverse()
    metadata.pop()
    metadata.reverse()
```
